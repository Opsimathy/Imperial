Standardise the data to prepare for PCA. 

```python
def standardise(X):
    """
    Normalise the data matrix X
    
    Arguments:
        X: (n, p^2) array, the data
            n is total number of data points
            p is number of pixels in each dimension
    
    return:
        Xbar: (N, p^2) array, the normalised data
    """
    
    mu = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    
    # we do not scale if the std is 0, so convert 0s in std to 1s
    std_filled = std.copy()
    std_filled[std==0] = 1.
    Xbar = ((X-mu)/std_filled)
    return Xbar
```

PCA finds Eigenvalues and Eigenvectors for the covariance matrix $C := X^T X / (n-1)$ where $n$ is total number of data points in $X$. This will be implemented below

```python
from scipy.sparse import linalg

def pca_full(X):
    """
    Perform the full PCA on given data matrix
    i.e. keep all principal components
    
    Arguments:
        X: (n, p) the data
            n: number of data points
            p: number of features
        
    Returns:
        explained_variances: X: (n, ) np array, the proportion of variance explained by each eigenvalue
        eigenvectors: (p, n) np array, each column is a principal components
        eigenvalues: (n,) np array, the corresponding eigenvalues, ordered from largest to smallest
    """
    
    n, p = X.shape
    # create covariance matrix C
    C = X.T @ X / (n-1)
    
    full_eigenvalues, full_eigenvectors = np.linalg.eig(C)
    # cast to real
    full_eigenvalues.astype(np.float64)
    full_eigenvectors.astype(np.float64)
    # sorting the eigenvectors and eigenvalues from largest to smallest eigenvalue
    sorted_index = np.argsort(full_eigenvalues)[::-1]  # reversed the output so that it is descending
    full_eigenvalues = full_eigenvalues[sorted_index]
    full_eigenvectors = full_eigenvectors[:,sorted_index]
    
    total_variance = sum(full_eigenvalues).real
    explained_variances = np.zeros(p)
    explained_variances[0] = full_eigenvalues[0] / total_variance
    for i in range(1, p):
        # cumulative sum of fraction 
        explained_variances[i] = explained_variances[i-1] + full_eigenvalues[i] / total_variance 
    
    return explained_variances, full_eigenvectors, full_eigenvalues
```

Perform iterative algorithm proposed by Lee and Seung to perform NMF on the data matrix. But first the data matrix X must be normalised to range [0, 1]

NMF requires the data matrix to be non-negative, so use the min-max normalisation

```python
def normalize_nmf(X):
    """
    Normalise X so that min(X) = 0, max(X) = 1
    
    Arguments:
        X: (n, p) array, the data to be normalised
    
    Returns:
        X_norm: (n, p) array, the normalised version of X
    """
    
    X_norm = (X - np.min(X)) / (np.max(X) - np.min(X))
    return X_norm
```

```python
def nmf(X, m, max_iter=500, tol=200):
    """
    Perform NMF keeping m dimensions
    using chi-squared loss
    
    Arguments:
        X: (n, p) array, the data to be normalised
        m: int, dimensions of projected space 
        max_iter: int, maximum number of iterations allowed
        tol: float, the convergence criteria, default 200
            stop iteration when chi-squared loss < tol
    
    Returns:
        W: (n, m) array
        H: (m, p) array
        iter_num: int, total number of iterations taken
        loss_l: list of losses
    """
    n, p = X.shape
    # initialisation 
    W = np.random.rand(n, m)
    H = np.random.rand(m, p)
    
    loss_l = []

    for iter_num in range(max_iter):

        # compute the update on H
        H = H * (W.T @ X) / (W.T @ W @ H)

        # compute the update on W
        W = W * (X @ H.T) / (W @ H @ H.T)

        # compute the chi2 loss and append to list
        diff = X - W @ H
        chi2_loss = np.sum(X * (diff ** 2)) / n * p
        loss_l.append(chi2_loss)
        if chi2_loss < tol:
            return W, H, iter_num+1, loss_l
        
    return W, H, max_iter, loss_l
```

For PCA, the reconstruction is by projection to the space spanned by first $m$ PCA. Suppose $V_m$ contains the first $m$ principal components (in columns). If $X_{new}$ is the test data matrix (where each row contains an image), the projected matrix is $X_{new} V_m$. Then for reconstruction we need to multiply $V_m^{-1} := V_m^T$ to bring $X_{new} V_m$ back to $\mathbb{R}^m$, so reconstruction $X_{PCA} := X_{new} V_m V_m^T$. This is a linear combination of columns of $V_m$ (principal components) with coefficients from $X_{new}V_m$.  </br>

Similar methods can be applied to NMF. Suppose $X = W_m H_m$ is the $m$-dimensional NMF, where rows of $H_m$ are NMF components. So the reconstruction $X_{NMF} := X_{new} H_m^T (H_m^T)^T = X_{new} H_m^T H_m$.  Again, $X_{NMF}$ is a linear combination of rows of $H_m$ with coefficients in $X_{new} H_m^T$.

First write a function to compute MSE between true data X and estimate `X_est`, the squared error is averaged among all the images contained in the data-set.

```python
def MSE_images(X, X_est):
    """
    Compute MSE for an estimation of the data
    
    Arguments:
        X: (n, p) np array, the true data set
        X_est: (n, p) np array, the estimated data set
    
    Return:
        MSE: float, the mean squared error
    """
    
    err = (X - X_est) ** 2
    # calcualte squared error for each image
    SE = np.sum(err, axis=1)
    
    return np.mean(SE)
```

#### 1.3 Gaussian Mixture Models (20 marks)

1.3.1 </br>

Assume the distribution of data follows GMM with 10 components:
$$P(X = \boldsymbol{x}) = \sum^{10}_{k=1} \pi_k p_k(\boldsymbol{x}|\boldsymbol{\theta})$$
where each $p_k$ is a multivariate-variate Gaussian, $\pi_k$ are the cluster proportions for each component, i.e. $\pi_k = P(C^{(i)} = k)$ where $C^{(i)}$ is the cluster data $x^{(i)}$ belongs to. <br>

First build a class for GMM model, which includes all the parameters required. Except means, variances of each component, $\pi_k$, the weights $w_{ik}$ for each data point in each cluster are also stored. The initialisation for $\pi_k$ and $w_{ik}$ are both uniform, and initial clustering is randomly assigned so that initial cluster means and variances can be calculated. 

```python
class GMModel:
    """
    Define the class of Gaussian mixture models
    
    fields:
        dim: number of mixture components K
        phi: (K, ) np array, cluster proportions pi_k
        weights: (n, m) weights w_{ik} assigned for each data point
            n - number of data points
            m - number of features
        mu: (K, m) np array, mixture component means for each cluster
            m - number of features
        sigma: list of covariance matrices ((m, m) np array) for each component
    """
    
    def __init__(self, X, dim):
        """
        Initialises parameters through random split of the data
        
        Arguments:
            X: (n, m) numpy array, the data matrix
                n - number of data points
                m - number of features
            dim: specified number of mixture components K
        """
        
        self.dim = dim

        # initial cluster proportions given to each cluster are stored in phi
        # equal clustering proportions assigned
        self.phi = np.full(shape=self.dim, fill_value=1/self.dim) 

        # initial weights given to each data point wrt to each cluster
        # equal weights assigned
        self.weights = np.full(shape=X.shape, fill_value=1/self.dim)
        
        n,m = X.shape 
        # dataset is divided randomly into k parts of unequal sizes
        random_row = np.random.randint(low=0, high=n, size=self.dim)

        # initial value of mean of k Gaussians
        # taken exactly to be the chosen representative data point of each part
        self.mu = [  X[row_index,:] for row_index in random_row ]  

        # initial value of covariance matrix of k Gaussians
        # all set to cov of entry X
        self.sigma = [ np.cov(X.T) for _ in range(self.dim) ] 
```

Build a function computing cluster probabilities (quantities $r_{ik}(\boldsymbol{\theta})$ given in lecture notes) for soft clustering

```python
def cluster_probabilities(gmm, X):
    """
    Predicts probability of each data point with respect to each cluster
    
    Arguments:
        gmm: an instance of GMModel class
        X: (n, m) numpy array, the data matrix
    
    Return:
        weights: (n, K) np array, the cluster probabilities r_ik
    """
    
    n,m = X.shape 
    # Creates a n*k matrix denoting likelihood belonging to each cluster 
    likelihood = np.zeros( (n, gmm.dim) ) 
    for k in range(gmm.dim):
        # likelihood of data belonging to k-th cluster 
        distribution = multivariate_normal(mean=gmm.mu[k], cov=gmm.sigma[k])
        likelihood[:,k] = distribution.pdf(X) 

        numerator = likelihood * gmm.phi
        denominator = numerator.sum(axis=1)[:, np.newaxis]
        weights = numerator / denominator
    return weights
```

To turn soft-clustering to hard-clustering, we need to find
$$\text{argmax}_{k} r_{ik}(\boldsymbol{\theta})$$
for each data $x^{(i)}$

```python
def predict(gmm, X):
    """
    Performs hard clustering using MAP estimate
    
    Arguments:
        gmm: object of GMModel class
        X: (n, m) numpy array, the data matrix
        
    Return:
        (n,) np array: MAP estimates
    
    """
    weights = cluster_probabilities(gmm,X)
    return np.argmax(weights, axis=1) 
```

Define a function for PCA that only computes the required dimensions, and project the data onto the space spanned by those principal components.

```python
from scipy.sparse import linalg

def pca_function(X,k):
    """
    Perform PCA on given data matrix
    
    Arguments:
        X: (n, m) the data
            n: number of data points
            m: number of features
        k: int, number of principal components to keep
        
    Returns:
        X_pca: X: (n, m) np array, the projected data matrix
        eigenvectors: (m, k) np array, each column is a principal components
        eigenvalues: (k,) np array, the corresponding eigenvalues i.e. variance of each principal component
    """
    
    n, _ = X.shape
    # create covariance matrix C
    C = X.T @ X / (n-1)

    # compute eigenvalues and eigenvectors using the eigsh scipy function
    # it returns only the largest k eigenvalues (in ascending order) and the corresponding eigenvectors
    eigenvalues, eigenvectors = linalg.eigsh(C, k, which="LM", return_eigenvectors=True) 
    # columns of variable `eigenvectors` are eigenvectors

    # sorting the eigenvectors and eigenvalues from largest to smallest eigenvalue
    sorted_index = np.argsort(eigenvalues)[::-1]  # reversed the output so that it is descending
    eigenvalues=eigenvalues[sorted_index]
    eigenvectors=eigenvectors[:,sorted_index ]  # must sort eigenvectors correspondingly

    # transform our data (i.e. project onto the k largest principal components)
    X_pca = X @ eigenvectors

    return X_pca, eigenvectors, eigenvalues
```

EM algorithm as described in lecture notes will be used to train the parameters in `gmm_10` model. The following convergence criteria is used:
$$ \sum^K_{i=1} \|\mu^{j+1}_i - \mu^j_i\|^2 + \|\sigma^{j+1}_i - \sigma^j_i\|^2 < \epsilon$$
where $j$ is the current iteration number and $i$ is the component(cluster) label. 2-norm is used for vectors and Frobenius norm is used for matrices $\sigma$. </br>
i.e. stop iteration when the mean and variance of GMM does not have a significant update.

```python
from scipy.stats import multivariate_normal 

def fitStep(gmm,X):
    """
    Performs one iteration of EM algorithm in-place (directly modify gmm)
    i.e. update all parameters
    
    Arguments:
        gmm: instance of GMModel class
        X: (n, p) numpy array, the data matrix
            n - number of data points
            p - number of features
    """
    
    # E-Step: update weights and phi holding mu and sigma constant
    weights = cluster_probabilities(gmm, X)  # find r_{ik}
    gmm.phi = weights.mean(axis=0)
        
    # M-Step: update mu and sigma holding pi and weights constant
    for k in range(gmm.dim):
        weight = weights[:, [k]] # take out (r_{ik})_{i = 1, ..., n}, (n, 1) np array
        total_weight = weight.sum()  # sum of r_{ik} along i

        gmm.mu[k] = (X * weight).sum(axis=0) / total_weight  
        gmm.sigma[k] = np.cov(X.T, aweights=(weight/total_weight).reshape(-1), bias=True)
    
def EM(gmm, X, max_it=500, tol=1e-3):
    """
    Perform EM algorithm to train the GMM model
    gmm will be updated in-place
    
    Arguments:
        gmm: instance of GMModel class
        X: (n, p) numpy array, the data matrix
            n - number of data points
            p - number of features
        max_it: int, maximum number of iterations, by default 200
        tol: tolerance for convergence criteria, default 1e-3
            criteria: total difference(l2-norm) between the Gaussian mean and variance between two iterations across all clusters
            
    Returns: 
        n_iter: int, total number of iterations taken
        diff_l: list of Gaussian mean and variance differences
    """
    
    # record real-time mean and variance to find the difference between two iterations
    mean = gmm.mu.copy()
    var = gmm.sigma.copy()
    k = gmm.dim
    diff_l = []
    
    for n_iter in range(1, max_it+1):
        fitStep(gmm, X)
        diff = 0
        for i in range(k):
            diff += np.linalg.norm(gmm.mu[i] - mean[i]) + np.linalg.norm(gmm.sigma[i] - var[i])
        diff_l.append(diff)
        if diff < tol:
            return n_iter, diff_l
        # update
        mean = gmm.mu.copy()
        var = gmm.sigma.copy()
    
    print("Warning: Maximum iteration reached for EM")
    return max_it, diff_l
```

```python
# train the model
iter_num_EM, diff_l_EM = EM(gmm_10, X_train_5)
print(f'Number of iterations taken: {iter_num_EM}')
```

```python
plt.figure(figsize=(12,8))
plt.title('EM algorithm convergence')
plt.xlabel('iteration number')
plt.ylabel('loss')
plot(np.arange(1,iter_num_EM+1), diff_l_EM)
```

```python
def label_map(gmm, X, true_labels):
    """
    Find a way to match predicted labels to class labels(true digits)
    by taking the cluster label with the highest log-likelihood for each class 
    
    Arguments:
        gmm: (n,) numpy array, instance of GMM model, already trained from data
        X: (n, p) numpy array, the data matrix
        true_labels: (n,) numpy array, class labels, or ground truth
            labels must be consecutive integers starting from 0 to K-1
        
    Return:
        label_map: (n,) array, i'th entry indicates which cluster label corresponds to digit i
        log_liks: (K, K) np array, ith row stores log-likelihoods of clusters for digit i
    """
    
    K = gmm.dim
    label_map = np.zeros(K) 
    log_liks = np.zeros((K, K)) # ith row stores log-likelihood of cluster for digit i
 
    for i in range(10):
        for k in range(K):
            distribution = multivariate_normal(mean=gmm.mu[k], cov=gmm.sigma[k])
            liks = distribution.pdf(X[true_labels == i])
            # deal with zero likelihood, simply ignore
            liks[liks == 0] = 1
            # find the mean log-likelihood for images in each cluster
            log_liks[i, k] = np.mean(np.log(liks))
        
        # pick the largest log-likelihood
        label_map[i] = np.argmax(log_liks[i,:])
        
    return label_map, log_liks

label_map_10, log_liks_10 = label_map(gmm_10, X_train_5, train_labels[:1000])
print(f"digits: {np.arange(10)}")
print(f"labels: {label_map_10.astype(int)}")
```

Now write a function to transform the labels to the correct digits

```python
def label_transform(predict_labels, label_map):
    """
    Transform the labels according to given label map 
    
    Arguments:
        predict_labels: (n,) array, predicted labels
            labels must be consecutive integers starting from 0
        label_map: (n,) array, i'th entry indicates with true label the ith predicted label should be mapped to 
    
    Return:
        new_labels: (n,) array, the transformed predicted labels
    """
    
    # initialise
    new_labels = predict_labels.copy()
    
    for i in range(len(label_map)):
        new_labels[predict_labels == label_map[i]] = i
    
    return new_labels
```

#### 2.1 Clustering (15 marks)
2.1.1

In this section, K-mean algorithm will be implemented and tested on gene data X. The scheme of K-mean is: </br>
1. Give random initialisation of clusters (i.e. randomly assign each sample to a cluster) 
2. Find the centroids for all clusters
3. assign each data point to the closest centroid, so we have new clusters
4. repeat step 2, 3 until maximum iteration reached or the cluster assignment no longer changes 

The distance used in K-mean is chosen to be the Euclidean distance i.e. 
$$\|\boldsymbol{x} - \boldsymbol{y} \|_2 := \sqrt{\sum_i (x_i - y_i)^2}$$

```python
def kmean(X, k, init_labels=None, max_iter=50, print_conv=True):
    """
    Perform K-mean clustering on the data matrix X 
    
    Arguments:
        X: (N, p) np array, the data matrix
            N: total number of samples
            p: number of features
        k: int, number of clusters
        init_labels: (N,) np array, the initial labels, None by default
            if no initial labels given, randomly generate labels
        max_iter: int, maximum number of iterations allowed 
        print_conv: boolean, true by default
            if true, print proportion of labels changed for each iteration
    
    Returns:
        labels: (N, ) np array, cosisting of labels to the data
            labels are from 1 to k
    """
    N, p = X.shape
    
    if init_labels:
        labels = init_labels
    else:
        labels = np.random.randint(low=0, high=k, size=N)
    
    # find initial centroids
    centroids = np.zeros((k, p))  # each row is a centroid for a cluster
    for i in range(k):
        if (labels == i).any():  # avoid summing empty labels
            centroids[i] = np.mean(X[labels==i], axis=0) 
    
    for i in range(max_iter):
        # distances: between data points and centroids
        distances = np.array([np.linalg.norm(X - c, axis=1) for c in centroids])
        # new_labels: computed by finding centroid with minimal distance
        new_labels = np.argmin(distances, axis=0)

        if (labels==new_labels).all():   # labels unchanged
            labels = new_labels
            print(f'Total number of iterations: {i+1}')
            return labels
        else:  # labels changed
            if print_conv:
                # difference: percentage of changed labels
                difference = np.mean(labels!=new_labels)
                print('%4f%% labels changed' % (difference * 100))
            labels = new_labels
            for c in range(k):
                # update centroids by taking the mean over associated data points
                if (labels == c).any():
                    centroids[c] = np.mean(X[labels==c], axis=0)
    
    print(f"warning: maximum iteration {max_iter} reached")
    return labels
```

The following function finds the Calinski-Harabasz index to assess clustering quality

```python
def CH_index(X, k, labels):
    """
    Calculate the Calinski-Harabasz index for a clustering result
    
    Arguments:
        X: (N, p) np array, the data matrix
            N: total number of samples
            p: number of features
        k: int, number of clusters
        labels: (N, ) np array, cosisting of labels to the data
            labels are from 1 to k
    
    Return:
        (float): Calinski-Harabasz index
    """
    N, p = X.shape
    
    
    centroids = np.zeros((k, p))
    for c in range(k):
        if (labels == c).any():
            centroids[c] = np.mean(X[labels==c], axis=0)
    
    # centroid of all points
    total_centroid = np.mean(X, axis=0)
    
    # find the number of data points in each cluster
    cluster_sizes = np.array([np.sum(labels==c) for c in range(k)])
    
    # between cluster scatter matrix
    BCSM = np.sum(cluster_sizes * np.sum((centroids - total_centroid)**2, axis=1))
    # withint cluster scatter matrix
    WCSM =0 
    for c in range(k): # find scatter for cluster c
        WCSM += np.sum(np.linalg.norm(X[labels==c] - centroids[c], axis=1) ** 2)

    return (BCSM / (k - 1)) * ((N - k)  / WCSM)
```

Search for the best k by maximising the average Calinski-Harabasz index for 5 different initialisations. Note by the formulae, larger index indicates larger between-cluster scatter and smaller within-cluster scatter, which means better clustering quality. Range of $k$ search: [2, 3, ... ,20]. 

The difference in initialisations of labels is achieved by using `np.random.randint` in the function `kmean`. It can also be seen from the difference in iteration numbers for each trial. 

```python
np.random.seed(123)

# store C-H indices for each k
index_l = []
ks = np.arange(2, 21)

for k in ks:
    print(f"k = {k}")
    # indices_k used to store CH indices for 5 initialisations
    indices_k = []
    for i in range(5):
        labels = kmean(X, k, print_conv=False)
        indices_k.append(CH_index(X, k, labels))
    index_l.append(np.mean(indices_k))
```

```python
plt.figure(figsize=(12, 8))
plt.title("plot of Calinski-Harabasz index for various k",size=20)
plt.xlabel("k",size=15)
plt.ylabel("index",size=15)
plt.plot(ks, index_l)
```

From the plot of indices against $k$, it can be seen that clearly the index is peaked at $k = 4$ and there is no need for further search as index keep decreasing from $k=5$ to $k=20$. This optimal $k$ is close to the true number of clusters $5$ given by column Type of the data file. 

```python
best_k = ks[np.argmax(index_l)]
best_k
```

The following function finds homogeneity score

```python
def homogeneity(C, K):
    """
    Calculate homogeneity score
    
    Arguments:
        C: (N,) np array, class labels(true labels) for the data
        K: (N,) np array, cluster labels of the data
        
    Return: 
        score: float, homogeneity score
    """
    
    N = len(C)
    if len(K) != N:
        raise Exception("length of C, K must be the same")
    
    K_size = len(np.unique(K))
    C_size = len(np.unique(C))
    
    # build a (|C|, |K|) array A where Ack indicates number of data points in class c assigned to cluster k
    A = np.zeros((C_size, K_size))
    for c in range(C_size):
        for k in range(K_size):
            A[c, k] = np.sum(np.logical_and(K == k, C == c))
            
    # conditional entropy
    # change 0 to 1, so that log(0) is mapped to 0 
    normalised_A = A / np.sum(A, axis=0)
    normalised_A[np.where(normalised_A == 0)] = 1
    H_con = - np.sum((A / N) * np.log(normalised_A))
    # total entropy
    H = - np.sum([(np.sum(A[c, :]) / N) * np.log(np.sum(A[c, :]) / N) for c in range(C_size)])
    
    return 1 - H_con / H
```

```python
hom_score = homogeneity(types_num, labels)
hom_score
```

From the homogeneity score of around $0.665$, the k-mean clustering does not align very well with the original data. Because this value is not close to $1$, i.e. ratio of conditional entropy $H(C | K)$ to $H(C)$ is not small. This indicates disagreement between classes C and clusters K.
