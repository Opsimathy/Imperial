```python
#Import Networkx, Numpty & Matplotlib
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define an empty graph G
G = nx.Graph()
```

```python
#Documentation about G (note different functions are needed for directed and multi-loop graphs)
G?
```

```python
dir(G)
```

```python
#Add edge between nodes 1 and 2
G.add_edge(1,2)
```

```python
#Check what nodes exist
G.nodes()
```

```python
edges = [(3,5),(4,6),(2,5), (1,4), (6,1), (3,2)]
```

```python
G.add_edges_from(edges)
```

```python
G.nodes()
```

```python
plt.figure()
nx.draw(G)
```

```python
plt.figure()
nx.draw_networkx(G, with_labels=True)
```

```python
#Clustering coefficient (number of links between neighbours of a node/ max no. of links possible)
nx.clustering(G)
```

```python
nx.degree(G)
```

```python
#Degree distribution of graph
nx.degree_histogram(G)
#ith element gives how many nodes have degree i
```

```python
#Graph Generator:
#1000 nodes and probably of a link being placed between two distinct nodes 0.01
G1 = nx.gnp_random_graph(1000,0.01)
```

```python
G1.number_of_nodes()
```

```python
G1.number_of_edges()
```

```python
#p_k is fraction of nodes with degree k
#Finds how many with degree k
P = nx.degree_histogram(G1)
```

```python
P = np.array(P)
```

```python
p_k = P/1000
```

```python
plt.figure()
plt.plot(p_k, '.')
plt.xlabel('degree, k')
plt.ylabel('p_k')
```
